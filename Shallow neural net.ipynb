```python
import numpy as np
np.random.seed(42)
```

```python
import keras
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import SGD  #stochastic gradient descent
```

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()  
```

```python
x_train.shape  #60000 digit images of size 28 by 28 of grey scale
```

```python
y_train.shape #60,000 classifiers/labels for each image/data
```

```python
y_train[0:20] #first 20 labels associated with digit images
```

```python
x_test.shape #10,000 images for test
```

```python
x_test[0] #first pixel image - not binary ; (0-254 pixel representation)
```

```python
y_test[0]  #actual image label
```

```python
x_trainb = x_train.reshape(60000, 784).astype('float32') #convert into vector ; 28*28 = 784 and float type
x_testb = x_test.reshape(10000, 784).astype('float32')
```

```python
x_trainb /= 255  # convert images in binary format; 0 or 1 pixel rep
x_testb /= 255
x_testb[0]
```

```python
n_classes=10
y_train = keras.utils.to_categorical(y_train, n_classes) # associate 0-9 categories, that converts a number to 10 binary digits indicating number one out of ten.
y_test = keras.utils.to_categorical(y_test, n_classes)
y_test[0]
```

```python
model = Sequential()
model.add(Dense(64, activation='sigmoid', input_shape=(784,)))   #input layer; 64 hyperparam = 64 neurons; 
model.add(Dense(10, activation='softmax'))
```

```python
model.summary()   #50240 possibilities; (64 * 784)+ 64 weights gor each neuron  ;; 650 = (10*64)+10
```

```python
model.compile(loss="mean_squared_error", optimizer=SGD(lr=0.01), metrics=['accuracy'])
```

```python
model.fit(x_trainb, y_train, batch_size=128, epochs=50, verbose=1, validation_data=(x_testb, y_test)) # train the model with running 50 iterations and batch size 128 and not all data
```

```python
model.evaluate(x_testb, y_test)
```
