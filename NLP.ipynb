```python
#wordtotvek ;; Gutenberg open source words corpus;; contains 18 books
#s-gram (skip-gram)is used find context words of center/key word #cbow (cont bag of words) is used to find key/center word from context words
import nltk
from nltk import word_tokenize, sent_tokenize
import gensim
from gensim.models.word2vec import Word2Vec
from sklearn.manifold import TSNE #scikit learn ; T distributed stochastic Neighbour Embedding => reducing dimensionality
import pandas as pd
from bokeh.io import output_notebook, output_file # javascript visualization lib
from bokeh.plotting import show,figure,ColumnDataSource
```

```python
len(gutenberg.fileids())
```

```python
gutenberg.fileids()
```

```python
gberg_sent_tokens = sent_tokenize(gutenberg.raw())
```

```python
word_tokenize(gberg_sent_tokens[1])
```

```python
gberg_sents = gutenberg.sents()
```

```python
gberg_sents[0:5]
```

```python
len(gutenberg.raw())
```

```python
#size = dimension; sg = skip gram =1 = true, window size of the context words = 10 words in bothe left & right dirction of center work
#min_cnt = consider only words which are apearing 5 times in the corpus
# seed no helps initializing the network;; 42= no of hyperparameters => dimensions, 
#workers = no of cores
# creates multi dimensional space vector;; time, relationship, animal, articles, stopwords, similar meaning words, etc
model = Word2Vec(sentences=gberg_sents, size=64, sg=1, window=10, min_count=5, seed=42, workers=6)
```

```python
model.save('raw.gutenberg_models.w2v')
```

```python
model['dog']
```

```python
model.most_similar('day')
```

```python
model.most_similar('think')
```

```python
model.doesnt_match("mother father daughter dog".split())
```

```python
model.wv.vocab #word vector 
```

```python
len(model.wv.vocab)
```

```python
#to view the model dimensions use tsne; by converting it to 2D
tsne = TSNE(n_components=2, n_iter=1000)
X=model[model.wv.vocab] #input
X_2d=tsne.fit_transform(X)
X_2d[0:5]
```

```python
crds_df = pd.DataFrame(X_2d, columns=['x','y'])
crds_df['token'] = model.wv.vocab.keys()
```

```python
crds_df.head()
```

```python
ds = ColumnDataSource(crds_df)
print(ds.data)
```

```python
crds_df.plot.scatter('x','y', figsize=(8,8), marker='o', s=10, alpha=0.2)
```

```python
output_notebook()
subset_df = crds_df.sample(n=1000)
p = figure(plot_width=600, plot_height=600)
p.text(x=subset_df.x, y=subset_df.y,
       text=subset_df.token)
show(p)
```
